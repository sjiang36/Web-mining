```
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
import csv
from requests.exceptions import RequestException
from urllib import robotparser
import time
```

```
rp = robotparser.RobotFileParser()
rp.set_url("https://scholar.google.com/robots.txt")
rp.read()
```

```
def get_profiles(names):
  url = "https://scholar.google.com/citations?hl=en&view_op=search_authors&mauthors="
  profile_urls = []
  for name in names:
    auth_url = url + str(name) + '&btnG='
    r = requests.get(auth_url)
    soup = BeautifulSoup(r.content, 'html.parser')
    links = soup.find_all("a")
    links = str(soup)
    expression = re.compile('user=(.{12})')
    index = re.findall(expression, links)
    
    if not index:
      profile_url = 'NULL'
    else:
      profile_url = 'https://scholar.google.com/citations?hl=en&user=' + str(index[0])
    profile_urls.append(profile_url)
    print(profile_url)
    # time.sleep(1)
  return profile_urls
```

```
def write_csv(file_name, urls):
  with open(file_name, 'a', encoding='utf-8') as file:
    writer = csv.writer(file)
    for url in urls:
        writer.writerow([url])
```

```
df = pd.read_csv('SCORE_csv.csv')
names = df['author_first_CR'] + '+' + df['author_last_CR']
names1 = names[:400]
names2 = names[400:800]
names3 = names[800:1200]
names4 = names[1200:1600]
names5 = names[1600:2000]
names6 = names[2000:2400]
names7 = names[2400:2800]
names8 = names[2800:]
```
